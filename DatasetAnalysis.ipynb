```python
#Encode categorical data (job)
df['job'].unique()
df['job'].value_counts()

from sklearn.preprocessing import LabelEncoder
job_encoder=LabelEncoder()

df['job']=job_encoder.fit_transform(df['job'])
df.head(10)
```

```python
job_encoder.classes_ #represents position of the job category in the array
```

```python
#Encode other classes (marital)
marital_encoder=LabelEncoder()
marital_encoder.fit(df['marital'])
df['marital']=marital_encoder.fit_transform(df['marital'])

#(education)
education_encoder=LabelEncoder()
education_encoder.fit(df['education'])
df['education']=education_encoder.fit_transform(df['education'])

#(default)
default_encoder=LabelEncoder()
default_encoder.fit(df['default'])
df['default']=default_encoder.fit_transform(df['default'])

#(housing)
housing_encoder=LabelEncoder()
housing_encoder.fit(df['housing'])
df['housing']=housing_encoder.fit_transform(df['housing'])

#(loan)
loan_encoder=LabelEncoder()
loan_encoder.fit(df['loan'])
df['loan']=loan_encoder.fit_transform(df['loan'])

#(contact)
contact_encoder=LabelEncoder()
contact_encoder.fit(df['contact'])
df['contact']=contact_encoder.fit_transform(df['contact'])

#(month)
month_encoder=LabelEncoder()
month_encoder.fit(df['month'])
df['month']=month_encoder.fit_transform(df['month'])

#(poutcome)
poutcome_encoder=LabelEncoder()
poutcome_encoder.fit(df['poutcome'])
df['poutcome']=poutcome_encoder.fit_transform(df['poutcome'])

#(y)
y_encoder=LabelEncoder()
y_encoder.fit(df['y'])
df['y']=y_encoder.fit_transform(df['y'])

df.head(10)
```

```python
#Using mean to segment demographics

features = df[['age', 'balance', 'marital', 'education']] #To feature selection
from sklearn.cluster import KMeans #Clustering
kmeans = KMeans(n_clusters=4)
df['segment'] = kmeans.fit_predict(features)

segment_profile = df.groupby('segment').mean() #Analyze each segments characteristics by grouping the data
print(segment_profile)
```

```python
#Visualisation
df.plot(kind='scatter', x='age',y='marital', color='green')
plt.show()
```

```python
#Train machine learning model
X=df[['job','marital','default','loan','housing','contact','education','campaign','poutcome']]
```

```python
X
```

```python
Y=df['y']
```

```python
Y
```

```python
#To split
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=12)
```

```python
X_train
```

```python
#Train the machine learning

# Model 1 Linear Regression

from sklearn.linear_model import LinearRegression
lr_model=LinearRegression()

lr_model.fit(X_train,Y_train) #Train the model
y_pred=lr_model.predict(X_test)

#Evaluate the model
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(Y_test,y_pred)

import math
x=math.sqrt(mse)
mse
```

```python
#Accuracy of the lr model
score=lr_model.score(X_test,Y_test)
print(f'Model score:{score*100:.2f}%')
```

```python
prediction=lr_model.predict([[4, 2, 0, 1, 1, 0, 2, 2, 3]]) #Row2 in the list
prediction
```

```python
# Model 2 SVR Support Vector Machine


from sklearn.svm import SVR
svr_model=SVR()

svr_model.fit(X_train,Y_train) #Train the model
y_pred_svr=svr_model.predict(X_test)

#Evaluate the model
mse_svr= mean_squared_error
(Y_test,y_pred_svr)
print(f"Mean Squared Error: {mse}")
svr_score=svr_model.score(X_test,Y_test)

print(f'Model score:{svr_score*100:.2f}%')

```
